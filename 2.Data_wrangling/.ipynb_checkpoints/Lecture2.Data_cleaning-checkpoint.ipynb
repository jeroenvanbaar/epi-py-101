# 1. What Is Missing Data?

Missing data appears when values are not recorded or are corrupted. In Pandas:
- `np.nan` or `None` represents missing data
- `NaN` = Not a Number (even for strings or categories)

# 2. Checking for missing data

```python
import pandas as pd
import numpy as np
```

```python
# Example dataset
patients = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'age': [34, np.nan, 45, 29],
    'smoker': ['yes', 'no', np.nan, 'no'],
    'bmi': [25.3, 27.8, np.nan, 24.5]
})

# Check for missing values
print(patients.isna())

# Summarize
print(patients.isna().sum())
```

```python
# Framingham dataset
filename = '../Data/frmgham2.csv'
frame = pd.read_csv(filename)
frame.head()
```

```python
# Check for missing values
print(frame.isna()) # Too large!
```

```python
# Summarize total missings per column
print(frame.isna().sum())
# Lots of missing data for H and L DLC!
```

```python
frame[['HDLC','LDLC','TOTCHOL']]
```

```python
# How many people have at least one missing data point?
frame['missing_per_row'] = frame.iloc[:,1:].isna().sum(axis=1)
frame[['RANDID','missing_per_row']].groupby('RANDID').sum().reset_index().sort_values('missing_per_row',ascending=True)
```

# 3. Dropping rows or columns

```python
# Drop rows with ANY missing value
frame.dropna()
```

```python
# Drop rows only if ALL values are missing
frame.dropna(how='all')
```

```python
# Drop columns with ANY missing value
frame.dropna(axis=1)
```

# 4. Filling missing data

Fill with a value

```python
frame['TOTCHOL'] = frame['TOTCHOL'].fillna(99)
```

```python
frame.head(15)
```

Fill with mean, median, or mode

```python
frame['BMI'] = frame['BMI'].fillna(frame['BMI'].mean())
frame['CURSMOKE'] = frame['CURSMOKE'].fillna(frame['CURSMOKE'].mode()[0])
```

```python
frame.head()
```

Note: we do not do any complicated imputation here. For instance MICE (multiple imputation with chained equations). This requires statistical analysis at a level not relevant for this course.

# 5. Detecting duplicate data

```python
frame.duplicated().sum()
```

```python
# If any duplicates:
frame = frame.drop_duplicates()
# ...but always find out WHY you have duplicate data.
```
