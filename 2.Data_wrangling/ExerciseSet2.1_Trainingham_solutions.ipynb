## 3. Fix data types

Task:
* Convert variable `'SEX'` from float to integer. To do this, first round the values to 0 decimals using `.round()` and then convert using `.astype('Int64')`.
* Convert variable `'AGE'` from float to integer. This one does not need rounding, so just use `.astype()`.

Note: Using 'Int64' instead of 'int' allows for NaN values to be retained.

```python
train['SEX'].head() # Before
```

```python
# Your code here
train['SEX'] = train['SEX'].round(decimals=0).astype('Int64')
```

```python
train['SEX'].head() # After
```

```python

train['AGE'] = train['AGE'].astype('Int64')
```

## 4. Missings and duplicates

**Tasks:**
* Find out how many missing values we have per row and column, and decide how to deal with them
* Find out how many duplicate rows there are, and fix them

```python
# Missings per column
missings_per_column = train.isna().sum()
missings_per_column
```

```python
# Missings per row
missings_per_row = train.isna().sum(axis=1).sort_values()
missings_per_row.head()
# Some people have no missing values this time! Why is that, you think?
```

```python
# Option of dealing with missings by removing columns with > 1000 missings
worst_columns = missings_per_column[missings_per_column > 1000].index.tolist()
train = train.drop(columns = worst_columns)
train.shape
```

```python
# Detect duplicates
train.duplicated().sum() # There are 581 duplicate rows!
```

```python
# Drop duplicates
train = train.drop_duplicates()
train.shape
```

## 5. Investigate Subgroups

**Tasks:**
- Select only female participants (`SEX == 2`)
- Calculate the average `AGE` of male and female participants separately (across all observations)
- Create a new DataFrame with only participants over age 60 who are current smokers

```python
# Females only
females_only = train.loc[train['SEX']==2]
females_only.head()
```

```python
# Average age for males and females
avg_age_male = train.loc[train['SEX']==1,'AGE'].mean()
avg_age_female = train.loc[train['SEX']==2,'AGE'].mean()
print(avg_age_female, avg_age_male)
```

```python
# Current smokers over 60
cursmoke_60plus = train.loc[(train['CURSMOKE']==1) & (train['AGE']>60),:]
cursmoke_60plus.head()
```
