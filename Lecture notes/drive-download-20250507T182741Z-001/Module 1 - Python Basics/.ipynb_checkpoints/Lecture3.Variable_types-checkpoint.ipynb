# Collections of data

```python
# List
list_of_integers = [3, 5,     7,2, 10]
print(list_of_integers)
```

```python
list_with_different_types = [3, 2.2, 'hi, my name is Jeroen']
print(list_with_different_types)
```

```python
# Access the contents of a list, you use an INDEX
print(list_with_different_types[1])
```

```python
print(type(list_with_different_types))
print(type(list_with_different_types[1]))
```

```python
# Replacing items
list_with_different_types[0] = 7
print(list_with_different_types)
```

```python
# Tuple
tuple_of_integers = (3, 5, 7)
print(tuple_of_integers)
```

```python
# Dictionary
dictionary_millions = {'USA':340,
                       'UK':68,
                       'NL':18,
                       'CA':40}
```

```python
print(dictionary_millions['CA'])
```

# Advanced printing

```python
# Print formatted strings with variable values in them
name = 'Jeroen'
age = 35
print(name,age)
print('My name is %s and my age is %i'%(name, age))
```

# Loop
Loop over values in a list

```python
for element in list_of_integers:
  print(element)
  print(element * 2 - 1)
```

```python
# Looping over dictionary
for name in dictionary_millions:
  print('Country %s has a population of %i million'%(name, dictionary_millions[name]))
```

# Logical operations and boolean variables

```python
print(3 > 2)
```

```python
print (5 == 10/2)
```

```python
print(3 < 2)
```

```python
# Print the country names of countries with more than 30 million people
for name in dictionary_millions:
  if dictionary_millions[name] > 30:
    print(name)
  else:
    print('Skipped one!')
```

```python
temp_today = 95
temp_yesterday = 93
# If temp is above 90 two days in a row, there is a heatwave
heat_wave = (temp_today > 90) & (temp_yesterday > 90)
print(heat_wave)
minor_heat_wave = (temp_today > 90) | (temp_yesterday > 90)
print(minor_heat_wave)
```
