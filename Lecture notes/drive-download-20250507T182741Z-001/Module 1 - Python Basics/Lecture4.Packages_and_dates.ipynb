# Pandas dataframes

```python
my_data = {
    'name':['Alice','Bob','Carlos'],
    'age':[34, 45, 58],
    'diagnosis-date':['2023-01-03','2023-01-04','2023-01-10']
}
print(my_data)

# Convert this to dataframe
patients = pd.DataFrame(my_data)
print(patients)
```

```python
type(patients)
```

```python
# Another Pandas application: dates
```

```python
patients.head()
```

```python
# Convert diagnosis date column to DATETIME format
patients['diagnosis-date'] = pd.to_datetime(patients['diagnosis-date'])
patients.dtypes
```

```python
# Extract the month
patients['month'] = patients['diagnosis-date'].dt.month
patients.head()
```

```python
# Filter patients diagnosed after Janaury 5th
patients[patients['diagnosis-date'] > '2023-01-05']
```
