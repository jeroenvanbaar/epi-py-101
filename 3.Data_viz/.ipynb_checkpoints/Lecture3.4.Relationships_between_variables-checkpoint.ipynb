# Lecture 4: Relationships Between Variables

### Objectives
- Visualize associations between two variables (categorical or continuous)
- Use scatterplots, trend lines, and correlation heatmaps
- Use encodings (color, size, shape) to show more dimensions
- Interpret grouped distributions using strip, box, and violin plots

## Load Framingham Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
df = pd.read_csv("../Data/frmgham2.csv")
```

## Relationship between two continuous variables
Visualize the relationship between BMI and systolic blood pressure (SYSBP)

```python
sns.scatterplot(data = df, x = 'BMI', y = 'SYSBP', alpha = .2)
plt.title("BMI vs. Systolic Blood Pressure")
```

# Add context with marginal distribution plots

```python
sns.jointplot(data=df, x="BMI", y="SYSBP", kind="scatter", marginal_kws=dict(bins=20, fill=True))
```

# Add a trend line (linear regression) using regplot

```python
sns.regplot(data=df, x="BMI", y="SYSBP",
            scatter_kws={"alpha":0.3, 's':1},
            line_kws = {'linewidth':2, 'linestyle':':'})
plt.title("BMI vs. Systolic Blood Pressure")
plt.show()
```

## Using `sns.lmplot` to Separate by Group
Plot the same relationship, but separated by sex (1 = male, 2 = female)

```python
sns.lmplot(data=df, x="BMI", y="SYSBP", hue="SEX", scatter_kws={"alpha":0.3, 's':1})
plt.title("BMI vs. SYSBP by Sex")
plt.show()
```

# Introduction to subplots

```python
# Matplotlib method
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = [10,4])
sns.regplot(data = df.query('CURSMOKE == 0'), x = 'BMI', y = 'SYSBP', ax = ax[0], scatter_kws = {'alpha':.2})
sns.regplot(data = df.query('CURSMOKE == 1'), x = 'BMI', y = 'SYSBP', ax = ax[1], scatter_kws = {'alpha':.2})
```

```python
# Seaborn method
sns.lmplot(data = df, x = 'BMI', y = 'SYSBP', col = 'CURSMOKE', hue = 'CURSMOKE',
           scatter_kws = {'alpha':.2})
```

# Correlation heatmap

```python
# Select relevant variables
subset = df[["BMI", "SYSBP", "DIABP", "AGE", "TOTCHOL"]].dropna()
corr = subset.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Health Variables")
plt.show()
```
