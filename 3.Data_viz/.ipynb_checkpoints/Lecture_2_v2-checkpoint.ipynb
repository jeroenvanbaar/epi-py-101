## Histograms and Kernel Density Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
tips = sns.load_dataset("tips")
tips.head()
```

```python
frame = pd.read_csv('../Data/frmgham2.csv')
frame.head()
```

# Research question 1: what is the distribution of age in this dataset?

* We are dealing with a single variable
* We want to count elements of that variable within bins
* => use histogram

```python
# Matplotlib version
plt.hist(frame['AGE'], bins = 10)
plt.xlabel('value')
plt.ylabel('count');
```

```python
# Seaborn version -- try this yourself!
sns.histplot(data = frame, x = 'AGE')
```

# Plotting smooth distribution shape: kernel density plot

```python
# KDE Plot
sns.kdeplot(data=frame, x="AGE", fill=True)
plt.title("KDE Plot of Age")
plt.show()
```

# Using log scales for skewed data

```python
# Matplotlib version
plt.hist(frame['TOTCHOL'], bins = 30)
plt.xlabel('value')
plt.ylabel('count');
plt.xscale('log')
```

```python
# Seaborn version
# KDE Plot
sns.kdeplot(data=frame, x="TOTCHOL", fill=True)
plt.xscale('log')
plt.title("KDE Plot of Total Cholesterol")
plt.show()
```

## Rug Plots and ECDFs

```python
# Powerful use of Seaborn: stacking plots
sns.kdeplot(data=frame, x="TOTCHOL", fill=True)
sns.rugplot(data=frame, x="TOTCHOL")
plt.title("KDE with Rug Plot")
plt.show()
```

```python
# ECDF Plot
sns.ecdfplot(data=frame, x="AGE")
plt.title("ECDF of AGE in Framingham Heart Study")
plt.show()
```
