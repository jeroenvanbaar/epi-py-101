## Introduction to `matplotlib.pyplot` and `seaborn`
Python has several plotting libraries. The most common in epidemiology are:
- `matplotlib.pyplot`: Low-level, very customizable
- `seaborn`: Built on top of matplotlib, easier for statistical graphics

We'll use both in this course.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set default plotting style
sns.set(style='whitegrid')
```

## Anatomy of a Plot
A typical plot includes:
- **Figure**: The whole window or page
- **Axes**: The area where data is plotted
- **Title**: What the plot is about
- **Axis Labels**: Explain what X and Y represent
- **Legend**: Explains colors/symbols if multiple series

We'll see this in action next.

```python
# Example: Manually constructing a plot with matplotlib
figure, axes = plt.subplots(nrows = 1, ncols = 1, figsize = [3,3]);
axes.plot([1, 2, 3], [1, 4, 9], label='Example')
axes.set_title('Sample Line Plot')
axes.set_xlabel('X-axis Label')
axes.set_ylabel('Y-axis Label')
axes.legend()
plt.show()
```

## The Role of Data Cleaning in Visualization
Visualization depends on clean and well-structured data. Watch out for:
- Missing values
- Incorrect types (e.g., strings instead of numbers)
- Redundant or duplicated rows

Clean data = more reliable plots.

```python
# Example dataset
raw_data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Cases': [120, 135, None, 160, 150, 170]
})
```

```python
fig, ax = plt.subplots()
ax.plot(raw_data['Month'], raw_data['Cases'])
```

```python
# Cleaning step: remove missing
cleaned_data = raw_data.dropna()
cleaned_data
```

```python
fig, ax = plt.subplots()
ax.plot(cleaned_data['Month'], cleaned_data['Cases'])
```

## Simple Bar and Line Plots using `matplotlib`

Instead of manipulating axes, we'll manipulate the entire plot using plt. functions

```python
# Scatter plot
plt.scatter(cleaned_data['Month'], cleaned_data['Cases'], marker='o')
plt.title('Monthly Disease Incidence')
plt.xlabel('Month')
plt.ylabel('Cases')
plt.show()
```

```python
# Line plot
plt.plot(cleaned_data['Month'], cleaned_data['Cases'], marker='o')
plt.title('Monthly Disease Incidence')
plt.xlabel('Month')
plt.ylabel('Cases')
plt.show()
```

```python
# Bar plot
plt.bar(cleaned_data['Month'], cleaned_data['Cases'])
plt.title('Monthly Disease Incidence')
plt.xlabel('Month')
plt.ylabel('Cases')
plt.show()
```

## First Plot with Seaborn

```python
# Load some example data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Matplotlib way of plotting total bill vs tips
fig, ax = plt.subplots(figsize = [3,3])
ax.scatter(tips['total_bill'], tips['tip'])
```

```python
# Seaborn way:
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

Key differences:
* Matplotlib is low-level. You explicitly define the figure, axes, and data points.
* Seaborn is high-level. You define the kind of graph you want, the data source, and which variables to map onto which axes. Seaborn fills in the required low-level commands.
* If Seaborn is a professor, Matplotlib is a grad student.
* If Seaborn is a general, Matplotlib is a soldier.
* If Seaborn is a CEO, Matplotlib is a product manager.
* And so on...

## Best Practices for Reproducibility and Labeling
- Always label axes and add a title
- Use legends if you have multiple groups
- Use consistent color schemes and themes
- Save your plots as files for reports and publications

```python
# Save plot as file
sns.histplot(tips['total_bill'])
plt.title('Distribution of Total Bill')
plt.savefig('histogram_total_bill.png', dpi=300) # Good choices: PNG, PDF, SVG. Bad choice: JPG.
plt.show()
```

# Storing plot in variable for later use

```python
fig,ax = plt.subplots()
sns.histplot(tips['total_bill'], ax = ax)
```
