### Objectives
- Compare distributions across subgroups using bar, box, violin, point, and swarm plots
- Understand error bars and confidence intervals in group comparisons

## Load the Framingham Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
frame = pd.read_csv("../Data/frmgham2.csv")
frame.head()
```

## Bar Plots: Comparing Group Means or Counts

```python
# Count of smokers vs non-smokers
sns.countplot(data=frame, x="CURSMOKE")
plt.title("Number of Smokers vs Non-Smokers")
plt.xlabel("Current Smoker (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Mean BMI by smoking status (CURSMOKE: 1 = smoker, 0 = non-smoker)
sns.barplot(data=frame, x="CURSMOKE", y="BMI", errorbar="sd")
plt.title("Average BMI by Smoking Status")
plt.xlabel("Current Smoker (0=No, 1=Yes)")
plt.ylabel("BMI")
plt.show()
```

## Box and Violin Plots: Showing Distributions by Group

```python
# Boxplot of TOTCHOL by sex
sns.boxplot(data=frame, x="SEX", y="TOTCHOL")
plt.title("TOTCHOL by Sex (1=Male, 2=Female)")
plt.xlabel("Sex")
plt.ylabel("TOTCHOL")
plt.show()
```

## Point and Swarm Plots: Good for Smaller Samples

```python
# Downsample for visibility
small_df = df.dropna(subset=['BMI', 'SEX', 'CURSMOKE']).sample(200, random_state=1)
sns.pointplot(data=small_df, x="SEX", y="BMI", hue="CURSMOKE", dodge=True, errorbar="sd")
plt.title("Mean BMI by Sex and Smoking Status")
plt.xlabel("Sex (1=Male, 2=Female)")
plt.ylabel("BMI")
plt.show()
```

```python
sns.swarmplot(data=small_df, x="SEX", y="BMI", hue="CURSMOKE")
plt.title("BMI per Observation by Sex and Smoking")
plt.xlabel("Sex (1=Male, 2=Female)")
plt.ylabel("BMI")
plt.show()
```

## Error Bars and Confidence Intervals
Many Seaborn plots compute bootstrapped 95% confidence intervals by default (can change with `errorbar=` argument).
- `errorbar="sd"` shows standard deviation instead
- Set `errorbar=None` to remove error bars

Use these to interpret uncertainty or variability.

# Putting it together

```python
# Combined mean bar with 95% CI + swarm
# Pause video and try yourself!
```

```python
sns.barplot(data=small_df, x="SEX", y="BMI", hue="CURSMOKE", errorbar=None, alpha = .5, zorder = 1)
sns.stripplot(data=small_df, x="SEX", y="BMI", hue="CURSMOKE", dodge=True, alpha = .8, zorder = 2, legend = None)
sns.barplot(data=small_df, x="SEX", y="BMI", hue="CURSMOKE", errorbar=('ci',95), capsize = .2, alpha = 0, zorder = 3, legend = None)
plt.title("BMI per Observation by Sex and Smoking")
plt.xlabel("Sex (1=Male, 2=Female)")
plt.ylabel("BMI")
plt.show()
```
